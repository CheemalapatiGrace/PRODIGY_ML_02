```python
import pandas as pd
mall_custo = pd.read_csv("Mall_Customers.csv")
mall_custo.head()
```

```python
X = mall_custo.iloc[:, [3, 4]].values
```

```python
import matplotlib.pyplot as plt
plt.scatter(mall_custo['Annual Income (k$)'],mall_custo['Spending Score (1-100)'],c="r")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
k = 5  
kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(x_scaled)
print(y_kmeans)
```

```python
import seaborn as sns

mall_custo['Cluster'] = y_kmeans
sns.scatterplot(data=mall_custo, x= mall_custo.columns[3], y=mall_custo.columns[4], hue='Cluster', palette='Set1')

plt.title('Customer Clusters')
plt.xlabel(mall_custo.columns[3])
plt.ylabel(mall_custo.columns[4])
plt.show()
```
